```python
## 1) Write a python function to compute the image of a fuzzy set A using extension principal given a mapping function f(x). Test the code for the following cases :

## a) A is a Triangular mapping function in range [1 to 10] and f(x) = x^4 + 3x^3
## b) A is a Trapezoidal mapping function in range [1 to 10] and f(x) = x^4 + 3x^3
## c) A is a Gaussian mapping function in range [1 to 10] and f(x) = x^4 + 3x^3
## d) A is a Generalized Bell mapping function in range [1 to 10] and f(x) = x^4 + 3x^3
## e) A is a Triangular mapping function in range [1 to 5] and f(x) = (x - 3)^2 + 2
## f) A is a Trapezoidal mapping function in range [1 to 5] and f(x) = (x - 3)^2 + 2
## g) A is a Gaussian mapping function in range [1 to 5] and f(x) = (x - 3)^2 + 2
## h) A is a Generalized Bell mapping function in range [1 to 5] and f(x) = (x - 3)^2 + 2
## i) A is a Triangular mapping function in range [0 to 8] and f(x) = sin(x)
## j) A is a Trapezoidal mapping function in range [0 to 8] and f(x) = sin(x)
## k) A is a Gaussian mapping function in range [0 to 8] and f(x) = sin(x)
## l) A is a Generalized Bell mapping function in range [0 to 8] and f(x) = sin(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the triangular membership function
def triangular_membership(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Define the trapezoidal membership function
def trapezoidal_membership(x, a, b, c, d):
    return np.maximum(np.minimum(np.minimum((x - a) / (b - a), 1), (d - x) / (d - c)), 0)

# Define the Gaussian membership function
def gaussian_membership(x, c, sigma):
    return np.exp(-((x - c) ** 2) / (2 * sigma ** 2))

# Define the Generalized Bell membership function
def bell_membership(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Define the function f1(x) = x^4 + 3x^3
def f1(x):
    return x**4 + 3*x**3

# Define the function f2(x) = (x - 3)^2 + 2
def f2(x):
    return (x - 3)**2 + 2

# Define the function f3(x) = sin(x)
def f3(x):
    return np.sin(x)

# Compute the image of the fuzzy set A using the extension principle
def compute_image_fuzzy_set(membership_function, x_range, params, f):
    x_values = np.linspace(x_range[0], x_range[1], 1000)
    membership_values = membership_function(x_values, *params)
    y_values = f(x_values)

    y_min, y_max = np.min(y_values), np.max(y_values)
    y_range = np.linspace(y_min, y_max, 1000)

    new_membership_values = np.zeros_like(y_range)

    for i, y in enumerate(y_range):
        mask = (y_values >= y - 1e-6) & (y_values <= y + 1e-6)
        if np.any(mask):
            new_membership_values[i] = np.max(membership_values[mask])

    return y_range, new_membership_values

# Plotting function
def plot_fuzzy_sets(ax, x1, membership1, label1, x2, membership2, label2, f_values, f_label):
    ax.plot(x1, membership1, color='red', label=label1)
    ax.plot(x2, membership2, color='blue', label=label2)
    ax.plot(x1, f_values, color='green', linestyle='--', label=f_label)
    ax.set_xlabel('x')
    ax.set_ylabel('Membership degree')
    ax.legend()
    ax.grid(True)

# Define the cases
cases = [
    (triangular_membership, [1, 10], [1, 5, 10], f1, 'Triangular', 'x^4 + 3x^3'),
    (trapezoidal_membership, [1, 10], [1, 4, 6, 10], f1, 'Trapezoidal', 'x^4 + 3x^3'),
    (gaussian_membership, [1, 10], [5, 2], f1, 'Gaussian', 'x^4 + 3x^3'),
    (bell_membership, [1, 10], [2, 4, 5], f1, 'Generalized Bell', 'x^4 + 3x^3'),
    (triangular_membership, [1, 5], [1, 3, 5], f2, 'Triangular', '(x - 3)^2 + 2'),
    (trapezoidal_membership, [1, 5], [1, 2, 4, 5], f2, 'Trapezoidal', '(x - 3)^2 + 2'),
    (gaussian_membership, [1, 5], [3, 1], f2, 'Gaussian', '(x - 3)^2 + 2'),
    (bell_membership, [1, 5], [1, 2, 3], f2, 'Generalized Bell', '(x - 3)^2 + 2'),
    (triangular_membership, [0, 8], [0, 4, 8], f3, 'Triangular', 'sin(x)'),
    (trapezoidal_membership, [0, 8], [0, 2, 6, 8], f3, 'Trapezoidal', 'sin(x)'),
    (gaussian_membership, [0, 8], [4, 2], f3, 'Gaussian', 'sin(x)'),
    (bell_membership, [0, 8], [2, 4, 4], f3, 'Generalized Bell', 'sin(x)')
]

# Plot the results
fig, axs = plt.subplots(6, 2, figsize=(15, 30))
axs = axs.flatten()

for i, (membership_function, x_range, params, f, label, f_label) in enumerate(cases):
    x_values = np.linspace(x_range[0], x_range[1], 1000)
    membership_values = membership_function(x_values, *params)
    y_values = f(x_values)
    y_range, new_membership_values = compute_image_fuzzy_set(membership_function, x_range, params, f)
    plot_fuzzy_sets(axs[i], x_values, membership_values, f'A ({label})', y_range, new_membership_values, f'B ({label} through {f_label})', y_values, f_label)

plt.tight_layout()
plt.show()
```

```python
# 2) Write a python function to compute the max-min composition of two fuzzy relation ?
```

```python
import numpy as np

def max_min_composition(R, S):
    """
    Compute the max-min composition of two fuzzy relations.

    Parameters:
    R (np.array): Fuzzy relation matrix R of size (m, n)
    S (np.array): Fuzzy relation matrix S of size (n, p)

    Returns:
    np.array: The max-min composition matrix T of size (m, p)
    """
    # Ensure the inner dimensions match
    assert R.shape[1] == S.shape[0], "Inner dimensions must match for matrix multiplication."

    m, n = R.shape
    n, p = S.shape

    # Initialize the composition matrix T with zeros
    T = np.zeros((m, p))

    # Compute the max-min composition
    for i in range(m):
        for j in range(p):
            T[i, j] = np.max(np.minimum(R[i, :], S[:, j]))

    return T

# Example usage
R = np.array([
    [0.1, 0.5, 0.3],
    [0.7, 0.2, 0.9]
])

S = np.array([
    [0.4, 0.8],
    [0.6, 0.3],
    [0.5, 0.7]
])

T = max_min_composition(R, S)
print("Max-Min Composition T:")
print(T)
```

```python
# 3) Write a python function to compute the max-product composition of two fuzzy relation ?
```

```python
import numpy as np

def max_product_composition(R, S):
    """
    Compute the max-product composition of two fuzzy relations.

    Parameters:
    R (np.array): Fuzzy relation matrix R of size (m, n)
    S (np.array): Fuzzy relation matrix S of size (n, p)

    Returns:
    np.array: The max-product composition matrix T of size (m, p)
    """
    # Ensure the inner dimensions match
    assert R.shape[1] == S.shape[0], "Inner dimensions must match for matrix multiplication."

    m, n = R.shape
    n, p = S.shape

    # Initialize the composition matrix T with zeros
    T = np.zeros((m, p))

    # Compute the max-product composition
    for i in range(m):
        for j in range(p):
            T[i, j] = np.max(R[i, :] * S[:, j])

    return T

# Example usage
R = np.array([
    [0.1, 0.5, 0.3],
    [0.7, 0.2, 0.9]
])

S = np.array([
    [0.4, 0.8],
    [0.6, 0.3],
    [0.5, 0.7]
])

T = max_product_composition(R, S)
print("Max-Product Composition T:")
print(T)
```

```python
# 4) Define two suitable primary linguistic terms (using mapping function or generalized bell mapping function) representing old and young people respectively over the age range [0 to 100]. Now obtain the membership functions for the following non-primary terms :

# a) Not very young and not very old.
# b) Very young or very old.
# c) Young but not very young.
# d) Extremely young or more or less old.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the generalized bell membership function
def bell_membership(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Define the age range
age_range = np.linspace(0, 100, 1000)

# Parameters for the bell functions for "young" and "old"
young_params = [20, 4, 20]  # [a, b, c] for young
old_params = [20, 4, 70]    # [a, b, c] for old

# Membership functions for "young" and "old"
young_membership = bell_membership(age_range, *young_params)
old_membership = bell_membership(age_range, *old_params)

# Plot the primary terms
plt.plot(age_range, young_membership, label='Young')
plt.plot(age_range, old_membership, label='Old')
plt.xlabel('Age')
plt.ylabel('Membership Degree')
plt.title('Primary Linguistic Terms')
plt.legend()
plt.grid(True)
plt.show()












# Define fuzzy operations
def not_very(term):
    return 1 - term ** 2

def very(term):
    return term ** 2

def extremely(term):
    return term ** 3

def more_or_less(term):
    return np.sqrt(term)

def intersection(term1, term2):
    return np.minimum(term1, term2)

def union(term1, term2):
    return np.maximum(term1, term2)

# Non-primary terms
not_very_young = not_very(young_membership)
not_very_old = not_very(old_membership)
very_young = very(young_membership)
very_old = very(old_membership)
young_but_not_very_young = intersection(young_membership, not_very_young)
extremely_young = extremely(young_membership)
more_or_less_old = more_or_less(old_membership)

# Combined terms
not_very_young_and_not_very_old = intersection(not_very_young, not_very_old)
very_young_or_very_old = union(very_young, very_old)
extremely_young_or_more_or_less_old = union(extremely_young, more_or_less_old)

# Plot the non-primary terms
fig, axs = plt.subplots(4, 1, figsize=(10, 20))

axs[0].plot(age_range, not_very_young_and_not_very_old, label='Not very young and not very old')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Membership Degree')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(age_range, very_young_or_very_old, label='Very young or very old')
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Membership Degree')
axs[1].legend()
axs[1].grid(True)

axs[2].plot(age_range, young_but_not_very_young, label='Young but not very young')
axs[2].set_xlabel('Age')
axs[2].set_ylabel('Membership Degree')
axs[2].legend()
axs[2].grid(True)

axs[3].plot(age_range, extremely_young_or_more_or_less_old, label='Extremely young or more or less old')
axs[3].set_xlabel('Age')
axs[3].set_ylabel('Membership Degree')
axs[3].legend()
axs[3].grid(True)

plt.tight_layout()
plt.show()
```

```python
# 5) Demonstrate the effect of contrast intensification on a fuzzy membership function.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the generalized bell membership function
def bell_membership(x, a, b, c):
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

# Define contrast intensification function
def contrast_intensification(mu):
    mu_prime = np.where(mu <= 0.5, 2 * mu**2, 1 - 2 * (1 - mu)**2)
    return mu_prime

# Define the age range
age_range = np.linspace(0, 100, 1000)

# Parameters for the bell functions for "young"
young_params = [20, 4, 20]  # [a, b, c] for young

# Membership functions for "young"
young_membership = bell_membership(age_range, *young_params)

# Apply contrast intensification
intensified_young_membership = contrast_intensification(young_membership)

# Plot the original and intensified membership functions
plt.plot(age_range, young_membership, label='Original Young', color='blue')
plt.plot(age_range, intensified_young_membership, label='Intensified Young', color='red', linestyle='--')
plt.xlabel('Age')
plt.ylabel('Membership Degree')
plt.title('Effect of Contrast Intensification on Young Membership Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a sample 1D membership function
def membership_function_1d(x):
    return np.exp(-((x - 50) ** 2) / (2 * (10 ** 2)))

# Define the cylindrical extension function
def cylindrical_extension(mu_1d, x_range, y_range):
    X, Y = np.meshgrid(x_range, y_range)
    Z = mu_1d(X)
    return X, Y, Z

# Define the x and y ranges
x_range = np.linspace(0, 100, 500)
y_range = np.linspace(0, 100, 500)

# Get the cylindrical extension
X, Y, Z = cylindrical_extension(membership_function_1d, x_range, y_range)

# Plot the cylindrical extension
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Membership Degree')
ax.set_title('Cylindrical Extension of 1D Membership Function')

plt.show()




# Define a sample 2D membership function
def membership_function_2d(x, y):
    return np.exp(-((x - 50) ** 2 + (y - 50) ** 2) / (2 * (10 ** 2)))

# Define the projection function
def projection_2d_to_1d(mu_2d, x_range, y_range):
    X, Y = np.meshgrid(x_range, y_range)
    Z = mu_2d(X, Y)
    projection = np.max(Z, axis=1)
    return x_range, projection

# Get the projection
x_proj, y_proj = projection_2d_to_1d(membership_function_2d, x_range, y_range)

# Plot the projection
plt.figure(figsize=(10, 6))
plt.plot(x_proj, y_proj, label='Projection of 2D Membership Function', color='red')
plt.xlabel('X')
plt.ylabel('Membership Degree')
plt.title('Projection of 2D Membership Function onto X-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 8) Given a fuzzy membership function mf(x) representing the fact x is A and fuzzy relation R(X,Y) representing the fuzzy implication if x is A then y is B, find the consequent membership function representing y is B. Implement this for both max-min and max-product compositions and demonstrate the process visually.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a sample fuzzy membership function mu_A(x)
def membership_function_A(x):
    return np.exp(-((x - 50) ** 2) / (2 * (10 ** 2)))

# Define a sample fuzzy relation R(x, y)
def fuzzy_relation_R(x, y):
    return np.exp(-((x - y) ** 2) / (2 * (15 ** 2)))

# Define the x and y ranges
x_range = np.linspace(0, 100, 100)
y_range = np.linspace(0, 100, 100)

# Compute the membership function mu_A(x)
mu_A = membership_function_A(x_range)

# Compute the fuzzy relation R(x, y)
X, Y = np.meshgrid(x_range, y_range)
R = fuzzy_relation_R(X, Y)

# Max-min composition
def max_min_composition(mu_A, R):
    result = np.zeros(R.shape[1])
    for j in range(R.shape[1]):
        result[j] = np.max(np.minimum(mu_A, R[:, j]))
    return result

# Max-product composition
def max_product_composition(mu_A, R):
    result = np.zeros(R.shape[1])
    for j in range(R.shape[1]):
        result[j] = np.max(mu_A * R[:, j])
    return result

# Compute the consequent membership functions
mu_B_max_min = max_min_composition(mu_A, R)
mu_B_max_product = max_product_composition(mu_A, R)

# Plot the results
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot mu_A(x)
axs[0].plot(x_range, mu_A, label='mu_A(x)', color='blue')
axs[0].set_title('Fuzzy Membership Function mu_A(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Membership Degree')
axs[0].legend()
axs[0].grid(True)

# Plot R(x, y)
cax = axs[1].imshow(R, extent=[0, 100, 0, 100], origin='lower', cmap='viridis', aspect='auto')
fig.colorbar(cax, ax=axs[1])
axs[1].set_title('Fuzzy Relation R(x, y)')
axs[1].set_xlabel('y')
axs[1].set_ylabel('x')

# Plot mu_B(y) for both compositions
axs[2].plot(y_range, mu_B_max_min, label='mu_B(y) - Max-Min Composition', color='green')
axs[2].plot(y_range, mu_B_max_product, label='mu_B(y) - Max-Product Composition', color='red', linestyle='--')
axs[2].set_title('Consequent Membership Function mu_B(y)')
axs[2].set_xlabel('y')
axs[2].set_ylabel('Membership Degree')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1) Write python functions to generate the following parameterized fuzzy membership functions and visualize them for different parameter values :

# a) Triangular MF
# b) Trapezoidal MF
# c) Gaussian MF
# d) Generalized Bell MF
# e) Sigmoidal MF
```

```python
import numpy as np
import matplotlib.pyplot as plt

def triangular_mf(x, a, b, c):
    """
    Generate a triangular fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        a (float): Left boundary.
        b (float): Peak/center.
        c (float): Right boundary.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

def trapezoidal_mf(x, a, b, c, d):
    """
    Generate a trapezoidal fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        a (float): Left shoulder.
        b (float): Left peak.
        c (float): Right peak.
        d (float): Right shoulder.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return np.maximum(0, np.minimum.reduce([((x - a) / (b - a)),
                                            np.ones_like(x),
                                            ((d - x) / (d - c))]))

def gaussian_mf(x, mean, sigma):
    """
    Generate a Gaussian fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        mean (float): Mean/center.
        sigma (float): Standard deviation.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)

def generalized_bell_mf(x, a, b, c):
    """
    Generate a generalized bell fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        a (float): Width parameter.
        b (float): Slope parameter.
        c (float): Center/mean.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return 1 / (1 + np.abs((x - c) / a) ** (2 * b))

def sigmoidal_mf(x, a, b):
    """
    Generate a sigmoidal fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        a (float): Steepness parameter.
        b (float): Midpoint parameter.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return 1 / (1 + np.exp(-a * (x - b)))

# Generate x values
x = np.linspace(0, 10, 1000)

# Visualize the membership functions for different parameter values
plt.figure(figsize=(12, 8))

# Triangular MF
plt.subplot(231)
plt.plot(x, triangular_mf(x, 2, 5, 8), label='Triangular MF')
plt.title('Triangular MF')
plt.legend()

# Trapezoidal MF
plt.subplot(232)
plt.plot(x, trapezoidal_mf(x, 1, 3, 7, 9), label='Trapezoidal MF')
plt.title('Trapezoidal MF')
plt.legend()

# Gaussian MF
plt.subplot(233)
plt.plot(x, gaussian_mf(x, 5, 1), label='Gaussian MF')
plt.title('Gaussian MF')
plt.legend()

# Generalized Bell MF
plt.subplot(234)
plt.plot(x, generalized_bell_mf(x, 1, 2, 5), label='Generalized Bell MF')
plt.title('Generalized Bell MF')
plt.legend()

# Sigmoidal MF
plt.subplot(235)
plt.plot(x, sigmoidal_mf(x, 2, 5), label='Sigmoidal MF')
plt.title('Sigmoidal MF')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 2) Write python function to generate asymmetric MF using sigmoidal MFs by the following methods and visualize them:

# a) Absolute difference
# b) Product
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoidal_mf(x, a, b):
    """
    Generate a sigmoidal fuzzy membership function.

    Parameters:
        x (numpy.ndarray): Input values.
        a (float): Steepness parameter.
        b (float): Midpoint parameter.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return 1 / (1 + np.exp(-a * (x - b)))

def asymmetric_abs_diff(x, a1, b1, a2, b2):
    """
    Generate an asymmetric fuzzy membership function using the absolute difference method.

    Parameters:
        x (numpy.ndarray): Input values.
        a1 (float): Steepness parameter for the left sigmoidal function.
        b1 (float): Midpoint parameter for the left sigmoidal function.
        a2 (float): Steepness parameter for the right sigmoidal function.
        b2 (float): Midpoint parameter for the right sigmoidal function.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return np.abs(sigmoidal_mf(x, a1, b1) - sigmoidal_mf(x, a2, b2))

def asymmetric_product(x, a1, b1, a2, b2):
    """
    Generate an asymmetric fuzzy membership function using the product method.

    Parameters:
        x (numpy.ndarray): Input values.
        a1 (float): Steepness parameter for the left sigmoidal function.
        b1 (float): Midpoint parameter for the left sigmoidal function.
        a2 (float): Steepness parameter for the right sigmoidal function.
        b2 (float): Midpoint parameter for the right sigmoidal function.

    Returns:
        numpy.ndarray: Fuzzy membership values.
    """
    return sigmoidal_mf(x, a1, b1) * sigmoidal_mf(x, a2, b2)

# Generate x values
x = np.linspace(0, 10, 1000)

# Visualize the asymmetric membership functions for different parameter values
plt.figure(figsize=(12, 5))

# Absolute difference method
plt.subplot(121)
plt.plot(x, asymmetric_abs_diff(x, 2, 3, 2, 7), label='Asymmetric MF (Abs Diff)')
plt.title('Asymmetric MF (Absolute Difference Method)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

# Product method
plt.subplot(122)
plt.plot(x, asymmetric_product(x, 2, 3, 2, 7), label='Asymmetric MF (Product)')
plt.title('Asymmetric MF (Product Method)')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 3) Write python function to implement following fuzzy complement operations and visualize them for different parameter values :

# a) Classical fuzzy complement
# b) Sugeno's fuzzy complement
# c) Yagar's fuzzy complement
```

```python
import numpy as np
import matplotlib.pyplot as plt

def classical_complement(membership_values):
    """
    Compute the classical fuzzy complement.

    Parameters:
        membership_values (numpy.ndarray): Fuzzy membership values.

    Returns:
        numpy.ndarray: Complement values.
    """
    return 1 - membership_values

def sugeno_complement(membership_values, alpha):
    """
    Compute Sugeno's fuzzy complement.

    Parameters:
        membership_values (numpy.ndarray): Fuzzy membership values.
        alpha (float): Complement parameter.

    Returns:
        numpy.ndarray: Complement values.
    """
    return np.exp(-alpha * membership_values)

def yager_complement(membership_values, alpha):
    """
    Compute Yager's fuzzy complement.

    Parameters:
        membership_values (numpy.ndarray): Fuzzy membership values.
        alpha (float): Complement parameter.

    Returns:
        numpy.ndarray: Complement values.
    """
    return np.minimum(1, np.maximum(0, 1 - membership_values ** alpha))

# Generate x values
x = np.linspace(0, 1, 100)

# Visualize the complement operations for different parameter values
plt.figure(figsize=(15, 5))

# Classical fuzzy complement
plt.subplot(131)
plt.plot(x, classical_complement(x), label='Classical Complement')
plt.title('Classical Fuzzy Complement')
plt.xlabel('Membership Degree')
plt.ylabel('Complement Degree')
plt.legend()

# Sugeno's fuzzy complement
plt.subplot(132)
alpha_sugeno = 2  # Adjust the parameter value
plt.plot(x, sugeno_complement(x, alpha_sugeno), label='Sugeno Complement')
plt.title('Sugeno\'s Fuzzy Complement ( = {})'.format(alpha_sugeno))
plt.xlabel('Membership Degree')
plt.ylabel('Complement Degree')
plt.legend()

# Yager's fuzzy complement
plt.subplot(133)
alpha_yager = 1.5  # Adjust the parameter value
plt.plot(x, yager_complement(x, alpha_yager), label='Yager Complement')
plt.title('Yager\'s Fuzzy Complement ( = {})'.format(alpha_yager))
plt.xlabel('Membership Degree')
plt.ylabel('Complement Degree')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 4) Write python function to implement following fuzzy intersection operation (T-norms) and visuals them for different parameter values:

# a) Minimum
# b) Algebraic product
# c) Bounded product
# d) Drastic product
```

```python
import numpy as np
import matplotlib.pyplot as plt

def minimum_tnorm(a, b):
    """
    Compute the minimum T-norm (fuzzy intersection) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.minimum(a, b)

def algebraic_product_tnorm(a, b):
    """
    Compute the algebraic product T-norm (fuzzy intersection) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return a * b

def bounded_product_tnorm(a, b):
    """
    Compute the bounded product T-norm (fuzzy intersection) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.maximum(0, a + b - 1)

def drastic_product_tnorm(a, b):
    """
    Compute the drastic product T-norm (fuzzy intersection) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.where(a == 1, b, np.where(b == 1, a, np.zeros_like(a)))

# Generate x values
x = np.linspace(0, 1, 100)

# Generate fuzzy sets for visualization
a = np.random.rand(100)
b = np.random.rand(100)

# Visualize the T-norms for different parameter values
plt.figure(figsize=(15, 5))

# Minimum T-norm
plt.subplot(141)
plt.plot(x, minimum_tnorm(a, b), label='Minimum T-norm')
plt.title('Minimum T-norm (Intersection)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Algebraic product T-norm
plt.subplot(142)
plt.plot(x, algebraic_product_tnorm(a, b), label='Algebraic Product T-norm')
plt.title('Algebraic Product T-norm (Intersection)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Bounded product T-norm
plt.subplot(143)
plt.plot(x, bounded_product_tnorm(a, b), label='Bounded Product T-norm')
plt.title('Bounded Product T-norm (Intersection)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Drastic product T-norm
plt.subplot(144)
plt.plot(x, drastic_product_tnorm(a, b), label='Drastic Product T-norm')
plt.title('Drastic Product T-norm (Intersection)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 5) Write python functions to implement following fuzzy union operation (S-norms) and visualize them for different parameter values:

# a) Minimum
# b) Algebraic Sum
# c) Bounded Sum
# d) Drastic Sum
```

```python
import numpy as np
import matplotlib.pyplot as plt

def minimum_snorm(a, b):
    """
    Compute the minimum S-norm (fuzzy union) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.maximum(a, b)

def algebraic_sum_snorm(a, b):
    """
    Compute the algebraic sum S-norm (fuzzy union) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return a + b - a * b

def bounded_sum_snorm(a, b):
    """
    Compute the bounded sum S-norm (fuzzy union) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.minimum(1, a + b)

def drastic_sum_snorm(a, b):
    """
    Compute the drastic sum S-norm (fuzzy union) of two fuzzy sets.

    Parameters:
        a (numpy.ndarray): First fuzzy set.
        b (numpy.ndarray): Second fuzzy set.

    Returns:
        numpy.ndarray: Resultant fuzzy set.
    """
    return np.where(a == 0, b, np.where(b == 0, a, np.ones_like(a)))

# Generate x values
x = np.linspace(0, 1, 100)

# Generate fuzzy sets for visualization
a = np.random.rand(100)
b = np.random.rand(100)

# Visualize the S-norms for different parameter values
plt.figure(figsize=(15, 5))

# Minimum S-norm
plt.subplot(141)
plt.plot(x, minimum_snorm(a, b), label='Minimum S-norm')
plt.title('Minimum S-norm (Union)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Algebraic sum S-norm
plt.subplot(142)
plt.plot(x, algebraic_sum_snorm(a, b), label='Algebraic Sum S-norm')
plt.title('Algebraic Sum S-norm (Union)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Bounded sum S-norm
plt.subplot(143)
plt.plot(x, bounded_sum_snorm(a, b), label='Bounded Sum S-norm')
plt.title('Bounded Sum S-norm (Union)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

# Drastic sum S-norm
plt.subplot(144)
plt.plot(x, drastic_sum_snorm(a, b), label='Drastic Sum S-norm')
plt.title('Drastic Sum S-norm (Union)')
plt.xlabel('Membership Degree')
plt.ylabel('Resultant Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```
